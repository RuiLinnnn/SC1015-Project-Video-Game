# Video Game

Predict which genre will earn the most sales in the future.

```
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```
vg_data = pd.read_csv('vgsales.csv')
print("Data dims : ", vg_data.shape)
vg_data.head()
```

```
# Information about the Variables
vg_data.info
```

# Data Preparation and Cleaning

```
# Remove data from 2017 - 2020, we only using 1980 - 2016 data
vg_data = vg_data[~vg_data["Year"].isin([2017, 2018, 2019, 2020])]
print("Data dims : ", vg_data.shape)
```

```
# checking for NA values
vg_data.isnull().sum()
```

```
# Drop the NA entries
vgresult=vg_data.dropna()
print("Data dims : ", vgresult.shape)
vgresult.head()
```

```
# check the clean dataset
vgresult.isnull().sum()
```

```
vgresult.to_csv('cleaned-Video-Game-dataset.csv')
```
